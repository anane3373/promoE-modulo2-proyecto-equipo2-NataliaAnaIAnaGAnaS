```python
def salario(sal):
    try:
        if sal in ['$0-999', '1,000-1,999', '2,000-2,999', '3,000-3,999', '4,000-4,999', '5,000-7,499', '7,500-9,999', '10,000-14,999', '15,000-19,999']:
            return '< 20K'
        elif sal in ['20,000-24,999','25,000-29,999', '30,000-39,999', '40,000-49,999']:
            return '20K - 50K'
        elif sal in ['50,000-59,999', '60,000-69,999', '70,000-79,999', '80,000-89,999', '90,000-99,999']:
            return '50K - 100K'
        elif sal in ['100,000-124,999', '125,000-149,999', '150,000-199,999']:
            return '100K - 200K'
        elif sal in ['200,000-249,999','250,000-299,999', '300,000-499,999','$500,000-999,999' , '>$1,000,000']:
            return '> 200K'
    except:
        return np.nan
```

```python
df_da['cat_salary'] = df_da['salary'].apply(salario)
df_da.head(2)
```

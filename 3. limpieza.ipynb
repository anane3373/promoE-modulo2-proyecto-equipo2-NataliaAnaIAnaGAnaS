```python
#miramos cuales son los rangos de coding_experience. Vamos a reorganizar estas categorías y a darles nombres más comprensibles.
df_adalab['coding_experience'].unique()
```

```python
def experience(experiencia):
    if experiencia == 'I have never written code':
        return 'Sin experiencia'
    elif experiencia == '< 1 years' or experiencia == '1-3 years':
        return 'Junior'
    elif experiencia == '3-5 years':
        return 'Mid'
    elif  experiencia == '5-10 years':
        return 'Senior'
    elif experiencia == '10-20 years' or experiencia == '20+ years':
        return 'Senior+'
```

```python
df_adalab['expererience_cut'] = df_adalab['coding_experience'].apply(experience)
df_adalab.head(2)
```

```python
#miramos cuales son los rangos de salary. Vamos a reorganizar estas categorías para no tener tantas y darles mayor sentido (salarios comunes en funcion de la experiencia).
df_adalab['salary'].unique()
```

```python
# Vamos a ver quién es el del salario atípico:
df_adalab[df_adalab['salary'] == '>$1,000,000']
```

```python
def salario(sal):
    try:
        if sal in ['$0-999', '1,000-1,999', '2,000-2,999', '3,000-3,999', '4,000-4,999', '5,000-7,499', '7,500-9,999', '10,000-14,999', '15,000-19,999', '20,000-24,999']:
            return '< 25K'
        elif sal in ['25,000-29,999', '30,000-39,999', '40,000-49,999']:
            return '25K - 50K'
        elif sal in ['50,000-59,999', '60,000-69,999', '70,000-79,999', '80,000-89,999', '90,000-99,999']:
            return '50K - 100K'
        elif sal in ['100,000-124,999', '125,000-149,999', '150,000-199,999', '>$1,000,000']:
            return '> 100K'
    except:
        return np.nan
```

```python
df_adalab['salary_cut'] = df_adalab['salary'].apply(salario)
df_adalab.head(2)
```

```python
#Eliminamos las columnas iniciales.
df_adalab.drop(['coding_experience', 'salary'], axis = 1, inplace = True)
```

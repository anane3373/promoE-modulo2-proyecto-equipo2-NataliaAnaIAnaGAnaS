```python
tree = ET.parse('../data/data_xml_clean.xml')
root = tree.getroot()

xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict)
```

```python
xml_series = pd.Series(xml_list) # creamos serie
```

```python
df_xml = pd.DataFrame.from_dict(xml_series)  # creamos DF malo a partir de los diccionarios
```

```python
para_limpiar_xml = df_xml[0].apply(pd.Series)  # se convierten k-v en filas-columnas
```

```python
para_limpiar_xml = para_limpiar_xml.drop(columns =["level_0"]) 
```

```python
para_limpiar_xml.set_index("index", inplace=True)  # cambiamos el indice
```

```python
para_limpiar_xml.to_csv('../data/df_xml.csv')  # guardamos en csv
```

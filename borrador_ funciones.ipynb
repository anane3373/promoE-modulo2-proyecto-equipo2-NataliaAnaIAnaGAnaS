```python
import pandas as pd
import numpy as np
```

```python
df_adalab = pd.read_csv('data/df_adalab.csv', index_col = 'index')
df_adalab.drop('Unnamed: 0', axis = 1, inplace = True)
df_adalab.head(2)
```

```python
class Limpieza:

    def __init__ (self, df):
        self.df = df

    def experience(experiencia):
        '''
        Recategorizamos el valor de la experiencia escribiendo código.
        Parámetro: 
            Experiencia: valor de la columna experiencia.
        Return:
            Nuevo valor de la categoría.
        '''
        if experiencia == 'I have never written code':
            return 'Sin experiencia'
        elif experiencia == '< 1 years' or experiencia == '1-3 years':
            return 'Junior'
        elif experiencia == '3-5 years':
            return 'Mid'
        elif  experiencia == '5-10 years':
            return 'Senior'
        elif experiencia == '10-20 years' or experiencia == '20+ years':
            return 'Senior+'

    def cambiar_col(self, col, funcion):
        self.funcion = funcion
        self.df[col] = self.df[col].apply(self.funcion)
        return self.df
```

```python
df_adalab_nuevo = Limpieza(df_adalab)
df_adalab_nuevo
```

```python
df_adalab_nuevo.cambiar_col('coding_experience', df_adalab_nuevo.experience())
```
